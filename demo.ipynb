```python
from azavea import City, Climate

c = Climate()
cities = c.city()
```

```python
cities
```

```python
points = []

for c in cities:
    points.append(tuple(c['geometry']['coordinates']))
```

```python
%matplotlib notebook

import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d

pts = np.array(points)

vor = Voronoi(pts)
```

```python
import matplotlib.pyplot as plt
voronoi_plot_2d(vor, show_points=True, show_vertices=False)
plt.show()
```
